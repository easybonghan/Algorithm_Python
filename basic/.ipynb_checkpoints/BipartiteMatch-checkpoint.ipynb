```python
def dfs(x):
    for node in adjacentList[x]:
        if isProcess[node]:
            continue
            
        isProcess[node] = True
        
        if usedNode[node] == 0 or dfs(usedNode[node]):
            usedNode[node] = x;
            return True
    
    return False

MAX          = 3
usedNode     = [0] * (MAX + 1)
isProcess    = [False] * (MAX + 1)
adjacentList = [[] for _ in range(MAX + 1)]
count     = 0

adjacentList[1].append(1)
adjacentList[1].append(2)
adjacentList[1].append(3)
adjacentList[2].append(1)
adjacentList[3].append(2)

for i in range(1, MAX + 1):
    isProcess = [False] * (MAX + 1)
    
    if dfs(i):
        count += 1

print(count)

for i in range(1, MAX + 1):
    if usedNode[i] != 0:
        print(usedNode[i], '->', i)
```
